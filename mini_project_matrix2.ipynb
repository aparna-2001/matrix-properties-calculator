```python
import csv
import zipfile
import io


zip_file_path = "archive.zip"
csv_filename_in_zip = "heart_cleveland_upload.csv"

try:
    with zipfile.ZipFile(zip_file_path, 'r') as zf:
        print(f"Files found inside '{zip_file_path}': {zf.namelist()}\n")


        if csv_filename_in_zip in zf.namelist():
            with zf.open(csv_filename_in_zip) as csv_file:
                # zf.open returns a file-like object in binary mode by default.
                # csv.reader expects a text file, so we wrap it in io.TextIOWrapper
                text_stream = io.TextIOWrapper(csv_file, encoding='utf-8') # Or the correct encoding of your CSV
                csv_rows = list(csv.reader(text_stream))
                print(f"Successfully read {len(csv_rows)} rows from '{csv_filename_in_zip}'.")

                # Printing the first few rows for demonstration
                if csv_rows:
                    print("\nFirst 5 rows:")
                    for i, row in enumerate(csv_rows[:5]):
                        print(row)
                    if len(csv_rows) > 5:
                        print("...")
                else:
                    print("The CSV file is empty.")
        else:
            print(f"Error: The CSV file '{csv_filename_in_zip}' was not found inside '{zip_file_path}'.")
            print(f"Please choose one of the files listed above.")

except zipfile.BadZipFile:
    print(f"Error: '{zip_file_path}' is not a valid zip file.")
except FileNotFoundError:
    print(f"Error: The file '{zip_file_path}' was not found. Please ensure it is uploaded or the path is correct.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

**matrix operations**

```python

rank = np.linalg.matrix_rank(data_matrix)
print(rank)
```
