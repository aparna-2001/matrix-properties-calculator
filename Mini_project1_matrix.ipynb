**CODE VERSION 1**

```python
#input of data

import numpy as np
import pandas as pd
import csv


#file input
data = pd.read_csv("data.csv") #replace it with your file name
print(data)

#checking the order
data_table = pd.DataFrame(data)
print(f"The data in table:{data_table}")

#conversion to matrix
def conversion_to_matrix(data):
    matrix_array = np.array(data)
    print(f"Converted matrix: {matrix_array}")
    return matrix_array

# Call the conversion function and store the result
matrix_input = conversion_to_matrix(data)


#finding the rank
def matrix_rank(matrix_input):
  matrix_rank = np.linalg.matrix_rank(matrix_input)
  print(f"Rank of matrix: {matrix_rank}")
  return matrix_rank

matrix_rank(matrix_input)



#finding the determinant
def det_matrix(matrix_input):
    try:
      matrix_1_det = np.linalg.det(matrix_input)
      print(f"Determinant of the matrix: {matrix_1_det}")
    except:
      print("This is a non square matrix")

det_matrix(matrix_input)

#finding the inverse
def inverse_matrix(matrix_input):
  try:
    matrix_inverse = np.linalg.inv(matrix_input)
    print(f"Inverse of the matrix: {matrix_inverse}")
  except:
    print("Inverse doesn't exist: The matrix is singular or the matrix is a non square matrix")

inverse_matrix(matrix_input)


#finding the Eigen values and Eigen vector
def eigen_matrix(matrix_input):
  try:
    Eigen_values,Eigen_matrix = np.linalg.eig(matrix_input)
    print(f"Eigen values of the matrix: {Eigen_values}")
    print(f"Eigen vectors of the matrix: {Eigen_matrix}")
  except:
    print("The matrix is not a square matrix")

eigen_matrix(matrix_input)
```
